```python
from keybert import KeyBERT
from transformers import pipeline
from bs4 import BeautifulSoup
import requests

# 1. Extrair texto de uma página web
url = 'https://pt.wikipedia.org/wiki/Brasil'
html = requests.get(url).text
text = BeautifulSoup(html, 'html.parser').get_text()

print(text)

```

```python
# 2. Extrair palavras-chave
kw_model = KeyBERT()
keywords = kw_model.extract_keywords(text, top_n=10)

# 3. Gerar resumo (abstrativo)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
summary = summarizer(text[:1024], max_length=150, min_length=30, do_sample=False)

print("Palavras-chave:", keywords)
print("Resumo:", summary[0]['summary_text'])
```
